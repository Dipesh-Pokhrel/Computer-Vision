TASK: Draw a BLUE Triangle in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orentation.

```python
vertices = np.array([[250,700],[425,400],[600,700]],np.int32)
vertices
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
pts
```

```python
cv2.polylines(img_rgb,[pts],isClosed=True,color=(0,0,255),thickness=10)
plt.imshow(img_rgb);
```
