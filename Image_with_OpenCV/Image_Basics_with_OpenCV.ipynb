So, the way matplotlib read image as RGB(Red, Green, Blue) channel. However OpenCV read as BGR (Blue, Green, Red) channel.

So, we transform Blue,Green,Red into RGB channel.

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img);
```

#### Resizing Images: 

```python
plt.imshow(img);
```

```python
fix_img.shape
```

```python
new_img = cv2.resize(fix_img,(900,500)) # Resizing images with height and width.
```

```python
plt.imshow(new_img);
```

We are able to resize by ratio instead.

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img1 = cv2.resize(fix_img,(0,0),fix_img,w_ratio, h_ratio)
plt.imshow(new_img1);
```

```python
new_img1.shape
```

Now, taking the 80% of the width ratio and 20% of the height ratio.

```python
h_ratio = 0.2
w_ratio = 0.8
new_img2 = cv2.resize(fix_img,(0,0),fix_img,w_ratio, h_ratio)
plt.imshow(new_img2);
```

Now, checking the shape of the array

```python
new_img2.shape
```

#### Flipping the images along the Horizontal and Vertical Axis.

```python
# Flipping image along the horizontal axis.
new_img3 = cv2.flip(fix_img,0)
plt.imshow(new_img3);
```

```python
# Flipping image along the vertical axis.
new_img3 = cv2.flip(fix_img,1)
plt.imshow(new_img3);
```
